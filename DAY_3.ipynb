````python
df['Final_Price'] = df['Total_Before_Discount'] * (1 - df['Discount_Pct'] / 100)
display(df['Final_Price'].head())
````

````python
total_revenue_by_category = df.groupby('Product_Category')['Final_Price'].sum().reset_index()
highest_revenue_category = total_revenue_by_category.loc[total_revenue_by_category['Final_Price'].idxmax()]
display(highest_revenue_category)
````

````python
total_price=df.groupby('Customer_Name')['Price'].sum().reset_index()
````

````python
top_3_customers = total_price.sort_values(by='Price', ascending=False).head(3)
display(top_3_customers)
````
